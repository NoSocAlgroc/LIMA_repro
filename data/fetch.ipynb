```python
import requests
import zipfile
import io
import os
```

```python
years=list(range(2020,2025))
months=list(range(1,13))

dataDir="data/"
os.makedirs(dataDir, exist_ok=True)
```

```python
for year in years:
    for month in months:
        fileName=f"On_Time_Reporting_Carrier_On_Time_Performance_1987_present_{year}_{month}.zip"
        url="https://transtats.bts.gov/PREZIP/"+fileName
        response = requests.get(url, stream=True)
        response.raise_for_status()  # stop if error

        if os.path.exists(dataDir+fileName):
            continue
        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            print(f"Extracting {url} -> {dataDir}")
            z.extractall(dataDir)
```

```python

output_file = "flights.csv"

with open(output_file, "w", encoding="utf-8") as fout:
    header_written = False

    for year in years[:1]:
        for month in months[:2]:
            fileName = f"On_Time_Reporting_Carrier_On_Time_Performance_(1987_present)_{year}_{month}.csv"
            path=dataDir+fileName
            if not os.path.exists(path):
                print(f"Skipping missing file: {fileName}")
                continue

            with open(path, "r", encoding="utf-8") as fin:
                # read header
                header = next(fin)

                # write header only once
                if not header_written:
                    fout.write(header)
                    header_written = True

                # write rest of the file
                for line in fin:
                    fout.write(line)
```
